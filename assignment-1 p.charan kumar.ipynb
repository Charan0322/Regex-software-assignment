```python
def pascal_triangle(size):
    for i in range(0, size):
        for j in range(0, i + 1):
            print(decide_number(i, j), end=" ")
        print()
                
                  
def decide_number(n,k):
    num = 1
    if k > n - k:
        k = n - k
    for i in range(0, k):
        num = num * (n-1)
        num = num // (i + 1)
    return num
    
                  
rows = 7
pascal_triangle(rows)
```
